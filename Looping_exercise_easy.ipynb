## **While loop**

* As long as the condition is met, the loop will continue to run.
* The loop will stop when the condition becomes False.

**Beware of infinite loop**

* Why? The program runs indefinitely because the condition is always True.  ---> MUST BE AVOIDED

```python
# Example of infinite loop
# x = 1
# while x <= 5:  # This will run indefinitely because x = 1 is always True  # MUST BE AVOIDED
#     print('I am learning Data Science')
#     # To stop: Press Ctrl + C
```

```python
# Example of while loop 1
x = 1
while x <= 5:
    print('I am learning Data Science')
    x = x + 1
```

```python
# Example of while loop 2
number = 1
while number <= 3:
    print(f'Number: {number}')
    number = number + 1
```

```python
# Example of a while loop 3  # Output: 2,3,4
number = 1
while number <= 3:
    number = number + 1
    print(f'Number: {number}')
```

**While Else**

## **For loop**

* The loop will continue to run as long as there are items or elements
inside a data collection or iterable.

**Looping over a Range (start, stop, step)**

* By default, the start value is 0, step size is 1,
* stop value is not included.

```python
# print('range(5)')
for value in range(5):  # range(stop): start=0, stop=5, step=1
    print(value)
```

```python
# print('range(2,5)')
for value in range(2,5):  # range(start, stop): start=2, stop=5, step=1
    print(value)
```

```python
# print('range(2,10,3)')
for value in range(2,10,3):  # range(start, stop, step): start=2, stop=10, step=3
    print(value)
```

```python
# print('range(3,0,-1)')
for value in range(3,0,-1):  # range(start, stop, step): start=3, stop=0, step=-1
    print(value)
```

**Looping through a Dictionary**

```python
# Display only the keys of dictionary
car_brands = {
    'toyota': 'innova',
    'honda': 'civic',
    'daihatsu' : 'sigra'
}


for brand in car_brands:
    print(brand)
```

```python
# Display only the values
for model in car_brands.values():
    print(model)
```

```python
# Display both keys and values
for brand, model in car_brands.items():
  print(f'Brand {brand} produces the {model} model')
```

```python
'''
# Task: Using a for loop on a dictionary, display:
# Output:
# Brand toyota produces the innova model
# Brand honda produces the civic model
# Brand daihatsu produces the sigra model
'''
for brand, model in car_brands.items():
    print(f'Brand {brand} produces the {model} model')
```

## **Break**

```python
# Break
text = 'I love to play guitar'
for char in text:
    if char == 'y':
        break
    print(char, end ='')
print()
```

```python
# Break
# When we are looking for "Juliana". After we have found her, we are stop looking.

friends = ["Dewi", "Diandra", "Lutfi", "Michael", "Juliana", "Fatimah", "Andi"]
for name in friends:
    if name == "Juliana":
        print("We have found Juliana!")
        break
    print("this is not Juliana:", name)
```

```python
# Stop when encountering a red stone
list_batu = ['yellow', 'green', 'purple', 'red', 'blue']
for batu in list_batu:
    if batu == 'red':
        print('Stop, there is a red stone, stop walking!')
        break  # Stop looping
    print(f'Stepping on the {batu} stone')
```

```python
'''
Question

Create a Python program that asks the user to input a password (for example, the password is '123').

The program must:

* Ask the user to enter the password repeatedly until it is correct.

* If the entered password is incorrect, display the message: "Incorrect password, please try again."

* If the entered password is correct, display the message: "Password correct" and stop the program.

Use a while True loop and a break statement.
'''

# Answer
password = 123

while True:
    pwd = int(input('Enter password: '))
    if pwd == password:
        print('Password correct')
        break
    print('Incorrect password, please try again.')
```

## **Continue**

```python
# Continue
for batu in list_batu:
    if batu == 'green':
        print('Green stone, just skip it')
        continue  # skip the green stone
    print(f'Stepping on the {batu} stone')
```

```python
# Greet all friends except "Lutfi"
friends = ["Dewi", "Diandra", "Lutfi", "Michael", "Juliana", "Fatimah", "Andi"]
for name in friends:
    if name == "Lutfi":
        continue  # Skip
    print("Hello", name)
```

**Nested loops**

* A nested loop is a loop inside another loop

How it works:
* The outer loop will run first
* Then the inner loop will run completely (finished) for each step of the outer loop


```python
classes = ["Data science", "Web Development", "Digital Marketing"]  # outer loop
for class_name in classes:
    print("Entering the class", class_name)
    for student in range(1, 3):  # student 1 and 2  # Inner loop
        print(f"  Student number {student} in class {class_name}")
```

```python
for bite in range(1,4):
    print(f'This is bite number {bite}')
    for dish in ['rice', 'beef stew', 'jackfruit leaves', 'chili sauce']:
        print(f'The dish is {dish}')
```
